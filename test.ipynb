```python
from google.cloud import bigquery
import pandas as pd
# import google.datalab.bigquery as bq

# def get_data_big(name):
#     table = 'bingo-ml-1.market_data.{}'.format(name)
#     fields = ['Date', 'Close']
#     data = bq.Query.from_table(bq.Table(table), fields=fields).execute().result().to_dataframe().set_index('Date')
#     return data['Close']


def get_data(name):
    client = bigquery.Client()
    query = "SELECT Date, Close FROM `bingo-ml-1.market_data.{}` ORDER BY Date".format(name)
    query_job = client.query(query)
    query_job.result() # Wait for result

    table = client.get_table(query_job.destination)
    rows = list(client.list_rows(table))
    return pd.Series([x[1] for x in rows], index=[x[0].date() for x in rows])


closing_data = pd.DataFrame()

# indexes = ['snp', 'nyse', 'djia', 'nikkei', 'hangseng', 'ftse', 'dax', 'aord']
indexes = ['snp', 'nyse', 'djia', 'nikkei', 'hangseng', 'ftse', 'dax', 'aord']

for index in indexes:
    closing_data['{}_close'.format(index)] = get_data(index)

closing_data = closing_data.fillna(method='ffill')

closing_data.describe()
```
